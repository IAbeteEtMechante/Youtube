```
from sklearn import datasets
```

```
iris = datasets.load_iris()
X = iris.data
y=iris.target
```

```
from sklearn.model_selection import train_test_split, GridSearchCV , RandomizedSearchCV
```

```
X_train, X_test, y_train , y_test = train_test_split(X,y, random_state = 69)
```

```
from sklearn.neighbors import KNeighborsClassifier
```

```
hyperparametres = [{'n_neighbors': range(1,31), 'weights': ['uniform']},{'n_neighbors': range(1,16), 'weights': ['uniform']},{'n_neighbors': range(1,31), 'weights': ['distance']}]
```

```
grid = GridSearchCV(KNeighborsClassifier(), hyperparametres, cv=10, scoring='accuracy')
```

```
grid.fit(X_train,y_train)
```

```
grid.best_params_
```

```
modele =grid.best_estimator_
```

```
modele.fit(X_train, y_train)
```

```
y_pred = modele.predict(X_test)
```

```
from sklearn.metrics import accuracy_score
```

```
accuracy_score(y_pred,y_test)
```

```
random_search = RandomizedSearchCV(KNeighborsClassifier(), hyperparametres, cv=10, n_iter= 10, scoring='accuracy')
```

```
random_search.fit(X_train,y_train)
```

```
random_search.best_params_
```

```
y_pred = random_search.predict(X_test)
```

```
accuracy_score(y_pred,y_test)
```
